```python
num_features = ['Age','Fare']
cat_features = ['Sex','Pclass','Family_Cat','Title', 'Embarked']
feature_cols = num_features + cat_features
print(feature_cols)
```

```python
def display_missing(df, feature_cols):
    n_rows = df.shape[0]
    for col in feature_cols: 
        missing_col = df[col].isnull().sum()
        if missing_col > 0:
            print(f"Col {col} has {missing_col*100/n_rows:.2f}% missing value")

display_missing(train_df, feature_cols)
display_missing(test_df, feature_cols)
```

```python
# pipeline
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
num_transformer = Pipeline(steps = [
    ('imputer', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])
```

```python
cat_transformer = Pipeline(steps = [
  ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown = 'ignore'))
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num',num_transformer, num_features),
    ('cat',cat_transformer, cat_features),
])
```

```python
preprocessor.fit(X)
```

```python
X = preprocessor.transform(X)
```

```python
X_test = preprocessor.transform(X_test)
```

```python
seed = 2023
models = [
    LinearSVC(max_iter=12000, random_state=seed),
    SVC(random_state=seed),
    KNeighborsClassifier(metric='minkowski', p=2),
    LogisticRegression(solver='liblinear', max_iter=1000),
    DecisionTreeClassifier(random_state=seed),
    RandomForestClassifier(random_state=seed),
    ExtraTreesClassifier(),
    AdaBoostClassifier(),
    XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=seed)
]
```

```python
from sklearn.model_selection import StratifiedKFold
def gen_baseline_results(models, X, Y, metrics, cv=5, plot_results=False):
    k_fold = StratifiedKFold(cv, shuffle=True, random_state=seed)
    entries = []
    for model in models:
        model_name = model.__class__.__name__
        score = cross_val_score(model, X, Y, scoring=metrics, cv=k_fold)
        for fold_idx, score in enumerate(score):
            entries.append((model_name, fold_idx, score))
            
    cv_df = pd.DataFrame(entries, columns=['model_name', 'fold_id', 'Accuracy_Scoring'])
    return cv_df

gen_baseline_results(models, X, Y, metrics='accuracy', cv=5, plot_results=False)
```
